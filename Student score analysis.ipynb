```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("expanded student scores.csv")
print(df.head())             
```

```python
#understanding the dataset
df.describe()
```

```python
#checking for discrepensies and understanding the data types
df.info()
```

```python
#checking for null values to clean the data
df.isnull().sum()
```

```python
#cleaning the data
```

```python
#dropping unnamed column
df = df.drop("Unnamed: 0", axis = 1)
df.head()
```

```python
#gender distribution chart
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender distribution")
plt.show()
```

Analysis: Number of females is greater than the number of males.

```python
#impact of parent education on the scores of the children
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parents Education and Student's Scores")
plt.show()
```

Analysis: The above heatmap concludes that the education of the parents does have an impact on the scores of the students. Parents with a higher degree show better scores and vice versa.

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parents Marital Status and Student's Scores")
plt.show()
```

Analysis: The impact of parents marital status on the scores of the students is negligible

```python
sns.boxplot(data = df, x = "MathScore")
plt.show()
```

```python
sns.boxplot(data = df, x = "ReadingScore")
plt.show()
```

```python
sns.boxplot(data = df, x = "WritingScore")
plt.show()
```

Analysis: After comparing the box plots, we can conclude that the students find Math to be harder than reading and writing due to increased outliers and reduced scores.
